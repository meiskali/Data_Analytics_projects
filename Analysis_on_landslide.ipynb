### Checking the null values in each columns in percent.

```python
for i in df1.columns:
    if df1[i].isnull().any():
        null=df1[i].isnull().sum()*100/df1.shape[0]
        print(i,"........",null,"%")
```

### Fill the NaN values according to their column behaviour.

```python
for i in df1.columns:
    if df1[i].isnull().sum()>0:
        print(i,".....",df1[i].dtypes)
```

```python
for i in df1.columns:
    if df1[i].isnull().sum()>0:
        if df1[i].dtypes=="float":
            df1[i].fillna(df1[i].median(),inplace=True)
        else:
            df1[i].fillna(df1[i].mode()[0],inplace=True)
```

### Now govt has relief scheme for landslide according to the landslide_size.


```python
df1["landslide_size"].unique()
```

#### Here we have problem that same size of landslide is defined in two way

```python
df1["landslide_size"]=df1["landslide_size"].str.replace("small","Small",case=False).str.replace("medium","Medium",case=False).str.replace("large","Large",case=False)
```

```python

df1["landslide_size"].unique()
```

#### Distribution of relief funds

```python
def condition(x):
    if x=="Small":
        return "1 Crore"
    elif x=="Medium":
        return "2 Crore"
    elif x=="Large":
        return "3 Crore"
    else:
        return "4 Crore"
```

```python
df1["Fund"]=df1["landslide_size"].apply(lambda x:condition(x))
```
